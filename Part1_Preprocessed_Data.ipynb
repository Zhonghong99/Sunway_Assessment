```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the **loyalty.csv** data

```python
df = pd.read_csv("/content/drive/My Drive/loyalty.csv")
df
```

Drop the column with null name

```python
drop_df = df.dropna(subset=["name"], axis = 0)
drop_df
```

Filter out the duplicated rows of data

```python
df.drop_duplicates(subset=None, keep=False, inplace=True)
```

```python
df
```

```python
selection = ["name","city"]
df2 = df[selection]
df2
```

```python
df2 = df2.replace(r'[^a-zA-Z\s]', '', regex=True)
df2.head(20)
```

Concatenate with the column **phone number**, **license plate**, **email** and **id**

```python
selected_col = ["phone-number","license-plate","email","id"]
df3 = pd.concat([df2,df[selected_col]], axis=1)
df3
```
